```python
import pylab
import itertools
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt

from load_batches_data import load_batches_data
```

```python
batches = load_batches_data("combined_batches_data.mat")
```

```python
batch = batches[0]

print("Variable names (keys): \n\n%s" % '\n'.join(batch.keys()))
```

```python
def plot_variable(ax, batch, variable, batch_num):
    ax.set_title("Batch no. %d, %s" % (batch_num, variable))
    ax.plot(batch[variable]["t"], batch[variable]["y"])
    ax.set_ylabel(batch[variable]["y_unit"])
    ax.set_xlabel(batch[variable]["t_unit"])
    
def plot_variables(num_batches, batches, variable):
    fig, ax = plt.subplots(ncols=num_batches, figsize=(20, 10))

    for idx, batch in enumerate(batches[:num_batches]):
        plot_variable(ax[idx], batch, variable, idx + 1)

    plt.show()
```

```python
%matplotlib inline

plot_variables(3, batches, "Base flow rate")
```
