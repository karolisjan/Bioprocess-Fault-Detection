```python
import numpy as np
import pandas as pd

from data_loader import load_batches_data
```

```python
batches = load_batches_data('Combined_batch_records1.mat')
```

```python
master = []

for i, batch in enumerate(batches['Combined_batch_records'].values()):
    df = {}
    
    for variable, values in batch.items():
        if not variable.lower().endswith('offline') and variable.lower() != 'stats':
            df['t'] = values['t'].tolist()
            df[variable] = values['y'].tolist()
    
    df = pd.DataFrame(df)
    df['batch_id'] = i + 1
    master.append(df)
    
master = pd.concat(master)
master
```

```python
master.to_csv('master.csv', index=False)
```
